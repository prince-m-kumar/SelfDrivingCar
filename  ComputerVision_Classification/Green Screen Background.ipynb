# Color Masking, Green Screen

```python
# for ploting the image
import matplotlib.pyplot as plt
# for reading the image
import matplotlib.image as mpimg
# for operating image and manuplate 
import numpy as np
import cv2

%matplotlib inline
```

```python
# Read in the image
image = mpimg.imread('images/car_green_screen.jpg')

# Print out the image dimensions (height, width, and depth (color))
print('Image dimensions:', image.shape)
# print output:- Image dimensions: (450, 660, 3) 450 pix:- height, 660 px:- width, 3:- number of color component have
# Display the image
plt.imshow(image)
```

```python
# Define our color selection boundaries in RGB values
lower_green = np.array([0,200,0]) 
upper_green = np.array([100,255,100])
```

#### Create a mask

```python
# Define the masked area
mask = cv2.inRange(image, lower_green, upper_green)
#print('Image dimensions:', mask.shape)
# Vizualize the mask
#plt.imshow(mask)
# Vizualize the mask
plt.imshow(mask, cmap='gray')
```

```python
# Mask the image to let the car show through
masked_image = np.copy(image)

masked_image[mask != 0] = [0, 0, 0]

# Display it!
plt.imshow(masked_image)
```
